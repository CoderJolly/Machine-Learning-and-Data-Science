## **Question - 0**
Null values are always a recurring part of any real-world dataset and it is important for a data scientist to deal with them. We analyse the number of NA values in our dataset.

```r
count_of_na <- apply(df, MARGIN = 2, function(x) sum(is.na(x)))
count_of_na_df <- data.frame(count_of_na)
count_of_na_df <- mutate(count_of_na_df,
                         columns = row.names(count_of_na_df),
                        )
```

```r
head(count_of_na_df, 20)
```

```r
# We will not plot those variables for which we dont have any NA values
options(repr.plot.width = 10, repr.plot.height = 10)
count_of_na_df %>%
filter(count_of_na > 0) %>%
ggplot(aes(reorder(columns, count_of_na), count_of_na))+
geom_bar(stat = "identity", fill = "tomato", color = "black")+
coord_flip()
```

## **Question - 1**
Let us consider the stamina of the players (stamina) for this question

### **1.1 Histogram**

```r
df %>%
ggplot(aes(stamina))+
geom_histogram(color = "red",fill = "pink", bins = 30)
```

### **1.2 Density**

```r
df %>%
ggplot(aes(stamina))+
geom_density(color = "red",fill = "pink")
```

### **1.3 Modality and Skewness**

Modality: From the above graph, we see that the distribution is bi-modal but with different modes. 

Skewness: We observe that the above distribution is left-skewed.

### **1.4 Mean, Variance, Standar Deviation and Skewness**

```r
mean <- mean(df$stamina, na.rm = TRUE)
standard_deviation <- sd(df$stamina, na.rm = TRUE)
variance <- var(df$stamina, na.rm = TRUE)
skewness <- skewness(df$stamina)

print(mean)
print(standard_deviation)
print(variance)
print(skewness)
```

### **1.5 Boxplots**

```r
boxplot(df$stamina, ylab = "Stamina of Players")
```

Let us calculate the statistics of boxplot.

```r
stamina_boxplot <- boxplot(df$stamina, plot = F)
stamina_boxplot$stats
```

We see from the above list that:

1. Q1 (25th percentile) = 56

2. Q3 (75th percentile) = 74

3. Median = 66

4. IQR = Q3-Q1 = 18

Lets calculate the fences:

1. Lower Inner Fence = Q1 - 1.5*IQR = 29

2. Lower Outer Fence = Q1 - 3*IQR = 2

3. Upper Inner Fence = Q3 + 1.5*IQR = 101

4. Upper Outer Fence = Q3 + 3*IQR = 128



### **1.6 Outliers in Boxplot**
Lets see how many outliers are present and where do they lie. From the graph, we can see that much that they all lie on the lower side of the data.

```r
stamina_boxplot$out
```

```r
length(stamina_boxplot$out)
```

There are total 827 outliers in the stamina distribution.
