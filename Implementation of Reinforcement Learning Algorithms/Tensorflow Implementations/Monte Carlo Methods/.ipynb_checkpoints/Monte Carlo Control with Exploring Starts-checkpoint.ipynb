## Monte Carlo Optimal Policy Control

```python
def mc_control_with_exploring_starts(policy, env, num_episodes, debug=False):
    
    returns_sum = defaultdict(float)
    returns_count = defaultdict(float)
    
    V = defaultdict(float)
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    final_policy = defaultdict(int)
    
    
    for i_episode in range(1, num_episodes+1):
        
        if debug:
            if i_episode % 100000 == 0:
                print("\rEpisode {}/{}.".format(i_episode, num_episodes))
                
        #Run an episode
        state = env.reset()
        start = True
        state_action_pairs_in_episode = []
        states_in_episode = []
        while(True):
            if start:
                action = np.random.choice([0,1])
                start=False
            else:
                action = final_policy[state]
            next_state, reward, end, _ = env.step(action)
            state_action_pairs_in_episode.append(((state, action),reward))
            states_in_episode.append(state)
            if end:
                break
            state = next_state
        
        for ((st,act), reward) in state_action_pairs_in_episode:
            first_occurence_idx = next(i for i,(s_a,r) in enumerate(state_action_pairs_in_episode) if s_a==(st,act))
            G = sum([r for ((s,a),r) in state_action_pairs_in_episode[first_occurence_idx:]])
            
            #Calculate average of the returns
            returns_sum[(st,act)] += G
            returns_count[(st,act)] += 1.0
            Q[st][act] = returns_sum[(st,act)] / returns_count[(st,act)]
            
            
        #Calculate the greedy policy
        for state_ in states_in_episode:
            curr_state_action_pairs = [(s,a) for ((s,a),r) in state_action_pairs_in_episode if s==state_]
            maximum=0
            for pair in curr_state_action_pairs:
                if Q[pair[0]][pair[1]]>maximum:
                    maximum=Q[pair[0]][pair[1]]
                    final_policy[state_] = pair[1]
                    
            
    return Q, final_policy

Q, optimal_policy = mc_control_with_exploring_starts(init_policy, env, num_episodes=500000, debug=True)
```
